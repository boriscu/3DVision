```python
import os
import shutil
from pathlib import Path
from PIL import Image
```

```python
def sample_images(input_dir, output_dir, step=1):
    """
    Samples every `step`-th image from input_dir and saves to output_dir.
    """
    input_path = Path(input_dir)
    output_path = Path(output_dir)
    output_path.mkdir(parents=True, exist_ok=True)

    images = sorted([f for f in input_path.iterdir() if f.suffix.lower() in ['.jpg', '.png']])
    sampled = images[::step]

    for img_path in sampled:
        shutil.copy(img_path, output_path / img_path.name)

    print(f"Copied {len(sampled)} images to {output_path}")
```

```python
gdrive_output_path = "/content/drive/My Drive/Posao/BioSense/Phenotyping/Data/FruitNeRF_Synthetic_Sampled_Step_1"
dataset_path = "/content/FruitNeRF_Synthetic/FruitNeRF_Synthetic/01_apple_tree_1024x1024_300/images"
```

```python
sample_images(dataset_path, gdrive_output_path, step=1)
```
